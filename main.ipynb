```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# Load dataset (update the path to your actual CSV file location)
df = pd.read_csv('C:/Users/Marc Jay/Ironhack/lab-eda-univariate/amz_uk_price_prediction_dataset.csv')

# Part 1: Understanding Product Categories

# Frequency table and top 10 categories
category_frequency = df['category'].value_counts()
top_10_categories = category_frequency.head(10)
other_sum = category_frequency.iloc[10:].sum()
top_10_categories['Other'] = other_sum

# Visualization: Bar chart for top categories
top_10_categories.plot(kind='bar', figsize=(12, 6))
plt.title('Top 10 Product Categories with "Other"')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

# Visualization: Pie chart for top categories with "Other"
top_10_categories.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Top 10 Categories Proportion with "Other"')
plt.ylabel('')
plt.show()

# Part 2: Delving into Product Pricing

# Centrality and dispersion for price
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]
variance_price = df['price'].var()
std_deviation_price = df['price'].std()
price_range = df['price'].max() - df['price'].min()
Q1_price = df['price'].quantile(0.25)
Q3_price = df['price'].quantile(0.75)
IQR_price = Q3_price - Q1_price

print(f"Price - Mean: {mean_price}, Median: {median_price}, Mode: {mode_price}")
print(f"Price - Variance: {variance_price}, Standard Deviation: {std_deviation_price}, Range: {price_range}, IQR: {IQR_price}")

# Visualization: Histogram and box plot for pricing
df['price'].plot(kind='hist', bins=100, figsize=(12, 6))
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 5))
df.boxplot(column='price')
plt.title('Price Spread and Outliers')
plt.ylabel('Price')
plt.show()

# Part 3: Unpacking Product Ratings

# Centrality and dispersion for ratings
mean_rating = df['stars'].mean()
median_rating = df['stars'].median()
mode_rating = df['stars'].mode()[0]
variance_rating = df['stars'].var()
std_deviation_rating = df['stars'].std()
IQR_rating = df['stars'].quantile(0.75) - df['stars'].quantile(0.25)

print(f"Rating - Mean: {mean_rating}, Median: {median_rating}, Mode: {mode_rating}")
print(f"Rating - Variance: {variance_rating}, Standard Deviation: {std_deviation_rating}, IQR: {IQR_rating}")

# Distribution shape for ratings
skewness = skew(df['stars'])
kurtosis_val = kurtosis(df['stars'])

print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis_val}")

# Visualization: Histogram for ratings
df['stars'].plot(kind='hist', bins=20, figsize=(12, 6))
plt.title('Product Ratings Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```
